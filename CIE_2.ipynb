```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

# Perform EDA
print("First 5 rows of the dataset:\n", tips.head())
print("\nDataset Info:")
print(tips.info())
print("\nSummary Statistics:\n", tips.describe())
print("\nColumns in dataset:", tips.columns)
print("\nUnique values in 'day':", tips['day'].unique())

# Find the missing values
print("\nMissing Values:\n", tips.isnull().sum())

# Scatter plot: Correlation between total_bill and tip
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Correlation between Total Bill and Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```
